Import Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
```

Defining the value of coeffient a and b

```python
a = 5
b = 3
c = 9
```

Plotting the linear function, y = ax + b

```python
x1 = np.linspace(-10, 10, 100)
y1 = a * x1 + b
plt.figure(figsize=(12, 6))

plt.plot(x1, y1, label=f'y={a}x+{b}')
plt.title(f'Plot of y = {a}x + {b}')
plt.xlabel('x')
plt.ylabel('y=f(x)')
plt.grid()
plt.legend()

plt.axhline(0, color='black', linewidth=2)
plt.axvline(0, color='black', linewidth=2)

plt.tight_layout()
plt.show()
```

Ptotting the quadratic function, y = ax^2 + bx + c

```python
x2 = np.linspace(-2, 2, 100)
y2 = a * (x2 ** 2) + b * x2 + c
plt.figure(figsize=(12, 6))

plt.plot(x2, y2, label=f'y={a}x+{b}')
plt.title(f'Plot of y = {a}x^2 + {b}')
plt.xlabel('x')
plt.ylabel('y=f(x)')
plt.grid()
plt.legend()

plt.axhline(0, color='black', linewidth=2)
plt.axvline(0, color='black', linewidth=2)

plt.tight_layout()
plt.show()
```

Plotting the trigonometric function, y = asin(x) + b

```python
x3 = np.linspace(-10, 10, 100)
y3 = a * np.sin(x3) + b
plt.figure(figsize=(12, 6))

plt.plot(x3, y3, label=f'y={a}sin(x)+{b}')
plt.title(f'Plot of y = {a}sin(x) + {b}')
plt.xlabel('x')
plt.ylabel('y=f(x)')
plt.grid()
plt.legend()

plt.axhline(0, color='black', linewidth=2)
plt.axvline(0, color='black', linewidth=2)

plt.tight_layout()
plt.show()
```

Plotting the exponential function, y = aexp(x) + b

```python
x4 = np.linspace(-3, 3, 100)
y4 = a * np.exp(x4) + b
plt.figure(figsize=(12, 6))

plt.plot(x4, y4, label=f'y={a}exp(x)+{b}')
plt.title(f'Plot of y = {a}exp(x) + {b}')
plt.xlabel('x')
plt.ylabel('y=f(x)')
plt.grid()
plt.legend()

plt.axhline(0, color='black', linewidth=2)
plt.axvline(0, color='black', linewidth=2)

plt.tight_layout()
plt.show()
```
